#### Exercício 1

```python
numero = int(input("Digite um número inteiro: "))
if numero % 5 == 0 and numero % 3 == 0:
    print("O número é divisível por 5 e por 3 ao mesmo tempo.")
else:
    print("O número não é divisível por 5 e por 3 ao mesmo tempo.")
```

#### Exercício 2

```python
valor = float(input("Digite o valor da compra: R$ "))
if valor <= 1000:
    desconto = 0.10
elif valor <= 5000:
    desconto = 0.20
else:
    desconto = 0.30
valor_desconto = valor * desconto
valor_final = valor - valor_desconto
print(f"Desconto: R$ {valor_desconto:.2f}")
print(f"Valor final: R$ {valor_final:.2f}")
```

#### Exercício 3

```python
largura = float(input("Digite a largura do quarto (m): "))
comprimento = float(input("Digite o comprimento do quarto (m): "))
altura = 2.80
porta_largura = 0.80
porta_altura = 2.10
area_paredes = 2 * (largura * altura) + 2 * (comprimento * altura)
area_porta = porta_largura * porta_altura
area_total = area_paredes - area_porta
litros_tinta = area_total / 3
print(f"Área total a pintar: {area_total:.2f} m²")
print(f"Quantidade de tinta necessária: {litros_tinta:.2f} litros")
```
