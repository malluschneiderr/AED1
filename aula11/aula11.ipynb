#### Exercício 9

```python
valor_hora = float(input("Digite o valor da hora trabalhada:"))
total_horas = float(input("Digite o total de horas trabalhadas:"))
sal_bruto = valor_hora * total_horas
fgts = sal_bruto * 0.11
sindicato = sal_bruto * 0.03

if sal_bruto <= 900 :
    a_ir = 0
elif sal_bruto <= 1500 :
    a_ir = 0.05
elif sal_bruto <= 2500:
    a_ir = 0.1
else :
    a_ir = 0.2
ir = sal_bruto * a_ir
sal_liquido = sal_bruto - (sindicato + ir)

print("Demonstrativo Salarial!")
print("-------------------------")
print("Salário Bruto: R$",sal_bruto)
print("Taxa Síndicato: R$",sindicato)
print("Desconto IR..: R$",ir)
print("Depósito FGTS: R$",fgts)
print("-------------------------")
print("Salário Liq..: R$", sal_liquido)
```
