#### Exercício 9

```python
valor_hora = float(input("Digite o valor da hora trabalhada:"))
total_horas = float(input("Digite o total de horas trabalhadas:"))
sal_bruto = valor_hora * total_horas
fgts = sal_bruto * 0.11
sindicato = sal_bruto * 0.03

if sal_bruto <= 900 :
    a_ir = 0
elif sal_bruto <= 1500 :
    a_ir = 0.05
elif sal_bruto <= 2500:
    a_ir = 0.1
else :
    a_ir = 0.2
ir = sal_bruto * a_ir
sal_liquido = sal_bruto - (sindicato + ir)

print("Demonstrativo Salarial!")
print("-------------------------")
print("Salário Bruto: R$",sal_bruto)
print("Taxa Síndicato: R$",sindicato)
print("Desconto IR..: R$",ir)
print("Depósito FGTS: R$",fgts)
print("-------------------------")
print("Salário Liq..: R$", sal_liquido)
```

```python
pop_a = 80000
pop_b = 200000

taxa_a = 0.03
taxa_b = 0.015

anos = 0

while pop_a < pop_b:
    pop_a = pop_a + (pop_a * taxa_a)
    pop_b = pop_b + (pop_b * taxa_b)
    anos += 1

print("Anos necessários:", anos)
print("População final A:", int(pop_a))
print("População final B:", int(pop_b))
```

#### Exercício 10

```python
soma = 0
quantidade = 0

maior = None
menor = None

while True:
    temp = float(input("Digite a temperatura (-1 para encerrar): "))
    
    if temp == -1:
        break

    soma += temp
    quantidade += 1

    if maior is None or temp > maior:
        maior = temp

    if menor is None or temp < menor:
        menor = temp

if quantidade > 0:
    media = soma / quantidade
    print("Maior temperatura:", maior)
    print("Menor temperatura:", menor)
    print("Média:", media)
else:
    print("Nenhuma temperatura foi informada.")
```
