### Exercício parte 1 
#### Passo 1 - Sortear um número aleatório entre 1 e 50
#### Passo 2 - Solicitar ao usuário um númeor entre 1 e 50
#### Passo 3 - Verificar se o número escolhido pelo usuário é igual ao sorteado 
#### Passo 4 - Se o número for igual, mostre a mensagem "Acertou! Parabéns!"
#### Passo 5 - Se o número for diferente, verifique se é maior ou menor, de a mensagem informando, retorne e solicite um outro número.

```python
import random 
n1 = random.randint(1,50)
while True :
    n2 = int(input("Digite o número da sorte:"))
    #print("O número sorteado foi: ",n1)
    if n2 == n1 :
        print("Acertou! Parabéns!!")
        break
    else:
        if n2 < n1 :
            print("Errou! O número escolhido é menor que o número correto. Tente outra vez!")
        else :
            print("Errou! O número escolhido é maior que o número correto. Tente outra vez!")
    n2 = n2 + 1
```

#### Exercício parte 2
#### Restringir para 5 tentativas 

```python
import random 
n1 = random.randint(1,50)
tentativa = 0
while True :
    if tentativa >= 5 :
        print("Suas tentativas acabaram!")
        break
    n2 = int(input("Digite o número da sorte:"))
    #print("O número sorteado foi: ",n1)
    if n2 == n1 :
        print("Acertou! Parabéns!!")
        break
    else:
        if n2 < n1 :
            print("Errou! O número escolhido é menor que o número correto. Tente outra vez!")
        else :
            print("Errou! O número escolhido é maior que o número correto. Tente outra vez!")
        tentativa += 1
```

#### Utiliando o for

```python
import random
n1 = random.randint(1,50)
for i in range(5) :
    n2 = int(input("Digite o número da sorte:"))
    if n2 == n1 :
        print("Acertou! Parabéns!!")
        break
    else:
        if n2 > n1 :
            print("Errou! O número escolhido é menor que o número correto. Tente outra vez!")
        else :
            print("Errou! O número escolhido é maior que o número correto. Tente outra vez!")
```

#### Utilizando quente e frio 

```python
import random
n1 = random.randint(1,50)
for i in range(5) :
    n2 = int(input("Digite o número da sorte:"))
    if n2 == n1 :
        print("Acertou! Parabéns!!")
        break
    else:
        if n2 < (n1 + 5) and n2 > (n1 - 5):
            print("Está quente!")
        else :
            print("Está frio!")
    
```
