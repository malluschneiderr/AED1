#### Exercício 1

#### Crie uma lista de nomes:
####  Nomes = ['Ana', 'Pedro', 'Bruno', 'Maria']
#### Em seguida, crie um loop de repetição, percorrendo a lista de nomes acima, e mostrando (print) um nome por linha.

```python
nome = ['Ana', 'Pedro', 'Bruno', 'Maria']
for i in range(4):
    print("Olá,",nome[i])
```

#### Outra forma de fazer

```python
alunos = ['Ana', 'Pedro', 'Bruno', 'Maria']
for nome in alunos: 
    print("Olá,",nome)
```

#### Exercício 2
#### Receba uma quantidade ilimitada (não se sabe quantas) de nome decidades e guarde essas cidades em uma lista.
#### Para encerrar a digitação das cidades, deve-se digitar: sair.
#### Em seguida, deve-se ordenar essa lista de cidades em ordem crescente.
#### E mostrar a relação de cidades ordenadas, 1 por linha.

```python
cidades = []
while True:
    cidade = input("Digite o nome da cidade (ou 'sair' para encerrar): ")
    if cidade.lower() == "sair":
        break
    cidades.append(cidade)
cidades.sort()
print("Cidades em ordem crescente:")
for cidade in cidades:
    print(cidade)
```
