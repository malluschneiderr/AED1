## Comandos Condicionais

#### If, Else e Elif 
#### Exemplo 1 

```python
idade = int(input("Digite sua idade: "))
if idade >= 18:
    print("Pode dirigir!")
else: 
    print("Chama o Uber!")
```

## Colocando mais Opções
#### Exemplo 2

```python
idade = int(input("Digite sua idade: "))
carro = input("Tem carro? <S/N> ")
if idade >= 18:
    if carro == "S":
        print("Pode dirigir!")
    else:
        print("Chama o Uber!")
else: 
    print("Chama o Uber!")
```

## Deixando o código menor
#### Exemplo 3

```python
idade = int(input("Digite sua idade: "))
carro = input("Tem carro? <S/N> ")
if (idade >= 18) and (carro == "S"):
    print("Pode dirigir!")
else: 
    print("Chama o Uber!")
```

### Exercício 1
#### Nota 1, Nota 2, fazer a média, resultado com false ou true

```python
nota1 = float(input("Digite sua primeira nota: "))
nota2 = float(input("Digite sua segunda nota: "))
media = (nota1 + nota2) / 2
print("A média é igual a: ", media)
if media >= 5 :
    print("Aprovado")
else :
    print("Reprovado")
```

## Exercício 2
#### Exercício anterior + Nota, nota >= 7 True (nota = nota + 0.5), false não faz nada
#### Tem dois jeitos de fazer, ambos demonstrados abaixo

```python
nota1 = float(input("Digite sua primeira nota: "))
if nota1 >= 7 :
    nota1 = nota1 + 0.5
nota2 = float(input("Digite sua segunda nota: "))
if nota2 >= 7 :
    nota2 += 0.5
media = (nota1 + nota2) / 2
print("A média é igual a: ", media)
if media >= 5 :
    print("Aprovado")
else :
    print("Reprovado")
```

## Exercício 3

#### 3 lados do trinângulo, verificar se os valores podem formar um triângulo, indicar caso que é, indicar qual é. 

```python
valor1 = float(input("De o valor do primeiro lado: "))
valor2 = float(input("De o valor do outro lado : "))
valor3 = float(input("De o valor da base: "))
soma = valor1 + valor2 
if (valor1 >= valor2 + valor3) or (valor2 >= valor1 + valor3) or (valor3 >= valor1 + valor2) : 
    print("Não é um triângulo")      
else : 
    if (valor1 == valor2 == valor3):
        print("Triângulo Equilátero.")
    elif (valor1 == valor2) or (valor2 == valor3) or (valor1 == valor3):
        print("Triângulo Isósceles.")
    else:
        print("Triângulo Escaleno.")
```
