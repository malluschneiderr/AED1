## While (Enquanto) loping de repetição

#### While condição:
#### - 
#### - 
#### -
#### Esse espaço (parágrafo) se chama identação. (Esse espaço é o bloco de código)
#### Exemplo (Mostra que todas ves que chaga na última linha o código retorna até ele se tornar <= 5)

```python
flag = 1
while flag <= 5:
    print ("Olá, sejam bem-vindos!")
    flag = flag + 1 
```

## For (variavel) in (interavel (Interagir) pode ser uma string, lista, qualquer coisa que tenha uma certa quantidade de elementos)

#### O comando range intera ( pode ter um inicio, passo e fim. Apenas o fim é obrigatório)

#### Exemplo: Gerar de 1 à 10

```python
#esse jeito não representa aonde quero chegar e sim aonde estou. Ou seja tenho que ir até o 11, porém não pego ele. 
#range é uma lista.
#for faz o consumo desses elementos
```

```python
for x in range (1,10):
    print(x)
```

```python
# pode também ser lista com nome
nome = input("Digite seu nome: ")
for letra in nome :
    print (letra)
```

```python
# Fazendo a repetição do while
for flag in range(1,6):
    print("Olá, sejam bem vindos!")
```

## Aplicando na prática 

```python
n = int(input("Digite o valor de N: "))
e = 0
for i in range(1,n+1):
    e = e + (2 ** i)

print ("O valor de E é igual: ", e)
```

```python
n = int(input("Digite o valor de N:"))
e = 0
i = 1 #inicializa com 1
while i <= n :
    print("e +", e , "i+",i)
    e = e + (2 ** i)
    i = i + 1

print("O valor de N: ", e)
```

```python
n = int(input("Digite a quantidade de pessoa:"))
soma_peso = 0 
soma_altura = 0
imc_maior = 0
imc_menor = 1000 
x = 1
while x <= n:
    peso = float(input("Digite o peso em kg:"))
    altura = float(input("Digite a altura em metros: "))
    imc = peso / (altura * altura)
    if imc > imc_maior:
        imc_maior = imc
    if imc < imc_menor:
        imc_menor = imc
    soma_peso += peso #mesma coisa que soma_peso = soma_peso + peso
    soma_altura += altura
    x += 1

media_peso = soma_peso / n
media_altura = soma_altura / n

print ("O imc mnaior =", imc_maior, "O imc Menor =", imc_menor)
print ("O peso médio é igual: ", media_peso)
print ("A altura média é igual: ", media_altura)
```

```python
#utilizando o comando continue 
#Explicação 1-10 não mostrar valores impares, números pares somar. 
soma = 0
n = 0
while n <= 10 :
    n =+ 1
    if (n % 2) == 1 :
        continue 
    soma += n #ou soma = soma + n
print ("Soma dos pares é igual: ", soma)
```

```python
#Exercício "Faça um programa que receba um número inteiro x. Calcule e mostre o fatorial desse número (x!)"
x = int(input("Digite o número de X: "))
if x == 0 :
    fatorial = 1
else:
    fatorial = 1
    i = 1
    while i <= x :
        fatorial = fatorial * i
        i += 1
print (x,"! = ",fatorial)
```
