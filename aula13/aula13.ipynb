```python
def soma(x,y):
    return x+y
```

```python
print("Título do Relatório")
mostrar_linha()
saudacao("Pedro")
print("O valor da Soma é:", soma(7, 5))
print("Descrição do Relatório bla bla bla...")
mostrar_linha()
```

#### Exercício 3

```python
def calcular_imc(peso,altura):
    imc = peso / (altura ** altura)
    return imc
```

```python
def classificar_imc(imc):
    if imc < 18.5 :
        return "Você está abaixo do peso"
    elif imc < 25 :
        return "Você está com o peso normal"
    elif imc < 30 :
        return "Você está sobrepeso"
    else: 
        return "Você está com Obesidade"
```

```python
peso = float(input("Digite seu peso:"))
altura = float(input("Digite sua altura:"))
imc = calcular_imc(peso,altura)
print (f"Seu IMC é: {imc:.2f} - {classificar_imc(imc)}")
```

#### Exercício 4

```python
def eh_primo(n):
    if n <= 1 :
        return False
    else:
        for i in range(2, n-1):
            if n % i == 0:
                return False
        return True
```

```python
while True :
    n = int(input("Digite um número primo:"))
    if eh_primo(n):
        print(f"{n} é um número primo!")
        break
    else :
        print(f"{n} não é um número primo! Tente Novamente!")
```
