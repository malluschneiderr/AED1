## Funções com Parâmetro Padrão

#### Exercício 1

#### Faça uma função que calcula o valor a ser cobrado pelos servições de entrega (losgítica) de uma loja na internet.
#### O custo é calculado tomando como base:
#### O peso do produto: até 500g -> 15,00 / até 2Kg -> 43,00 / acima de 2Kg -> 103,00 (reais)
#### O estado de destino: Norte/Nordeste -> 50,00 / Centro Oeste/Sul -> 25,00 / Sudeste -> 0,00 (reais)
#### Presente (é presente): Considerar sempre que não é presente = false / Se for presente (True) acrescentar 10,00 (reais)

```python
def regiao(estado):
    norte = ["AC", "AP", "AM", "PA", "RO", "RR", "TO"]
    nordeste = ["AL", "BA", "CE", "MA", "PB", "PE", "PI", "RN", "SE"]
    centro_oeste = ["DF", "GO", "MT", "MS"]
    sudeste = ["ES", "MG", "RJ", "SP"]
    sul = ["PR", "RS", "SC"]
    
    if estado.upper() in norte:
        return "N"
    if estado.upper() in nordeste:
        return "NE"
    if estado.upper() in centro_oeste:
        return "CO"
    if estado.upper() in sul:
        return "S"
    if estado.upper() in sudeste:
        return "SE"
```

```python
def frete(peso, estado, presente = False):
    '''
        DocString usando as 3 aspas simples ou duplas, para documentação e entendimento do código
    '''
    #calculo valor do peso 
    if peso <= 500:
        vp = 15
    elif peso <= 2000:
        vp = 43
    else :
        vp = 103

    #Calculo valor do estado
    if regiao(estado) == "NE" or regiao(estado) == "N" :
        ve = 50
    elif regiao(estado) == "CO" or região(estado) == "S" :
        ve = 25 
    else :
        ve = 0 

    #calculando valor do presente 
    if presente :
        vpr = 10
    else :
        vpr = 0

    return vp + ve + vpr
```

```python
peso = float(input("Qual o peso do produto?"))
estado = input("Qual o estado de entrega do produto?")
presente = input("Será para presente? (S/N)")

if presente in "Ss" :
    valor_frete = frete(peso, estado, True)
else: 
    valor_frete = frete(peso, estado, False) 

print("O valor total do frete é: ", valor_frete)
```
